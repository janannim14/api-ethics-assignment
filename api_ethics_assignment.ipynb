```python
import pandas as pd
import hashlib

data = {
    'full_name': ['Jananni', 'Arunpandi', 'Vijay'],
    'email': ['jananni@email.com', 'arun@email.com', 'vj@email.com'],
    'date_of_birth': ['1990-05-15', '1985-11-22', '1992-01-10'],
    'zip_code': ['600001', '600028', '641004'],
    'job_title': ['Data Analyst', 'Doctor', 'Engineer'],
    'diagnosis_notes': ['Patient has mild fever', 'Name: Meera, blood pressure high', 'Back pain']
}

df = pd.DataFrame(data)

df_cleaned = df.drop(columns=['email'])

df_cleaned['patient_id'] = df_cleaned['full_name'].apply(lambda x: hashlib.sha256(x.encode()).hexdigest()[:10])

df_cleaned['birth_year'] = pd.to_datetime(df_cleaned['date_of_birth']).dt.year
df_cleaned = df_cleaned.drop(columns=['date_of_birth'])

df_cleaned['zip_code'] = df_cleaned['zip_code'].str[:3] + '***'

print("Cleaned Dataset for Sharing:")
print(df_cleaned.head())
```

```python
import requests
import time
import os

API_URL = "https://healthstats-api.example.com/records"
API_KEY = os.getenv("API_KEY", "dummy_key_for_testing")

records = []

for page in range(1, 3):
    try:
        print(f"Fetching page {page}...")

        response = requests.get(API_URL, params={"page": page, "key": API_KEY}, timeout=5)
        response.raise_for_status()

        data = response.json()
        records.extend(data.get("results", []))

        time.sleep(1)

    except requests.exceptions.RequestException as e:
        print(f"Error on page {page}: URL doesn't exist (Example URL used).")
        print(f"Actual Error: {e}")

print("\nTask 2 Audit Complete: Code is now secure and ethical.")
```
